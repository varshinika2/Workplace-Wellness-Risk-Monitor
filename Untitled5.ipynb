```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('/content/healthcare_dataset.csv')

# Drop unnecessary columns for health monitoring
columns_to_use = ['Age', 'Medical Condition', 'Billing Amount', 'Admission Type', 'Medication', 'Test Results']
df = df[columns_to_use].copy()

# Encode categorical features
df['Medical Condition'] = df['Medical Condition'].astype('category').cat.codes
df['Admission Type'] = df['Admission Type'].astype('category').cat.codes
df['Medication'] = df['Medication'].astype('category').cat.codes
df['Test Results'] = df['Test Results'].map({'Normal': 0, 'Abnormal': 1, 'Inconclusive': 0.5})

# Drop missing values if any
df = df.dropna()

print("âœ… Data preprocessed successfully")
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
features = ['Age', 'Medical Condition', 'Billing Amount', 'Admission Type', 'Medication', 'Test Results']
df_scaled = scaler.fit_transform(df[features])
```

```python
from sklearn.ensemble import IsolationForest

iso = IsolationForest(contamination=0.05, random_state=42)
df['Anomaly'] = iso.fit_predict(df_scaled)
df['Risk Score'] = -iso.decision_function(df_scaled)

print("âœ… Anomaly detection complete")
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)
df['Risk Group'] = kmeans.fit_predict(df[['Risk Score']])

# Label risk groups
df['Risk Label'] = df['Risk Group'].map({0: 'Low Risk', 1: 'Medium Risk', 2: 'High Risk'})
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Risk Score', hue='Risk Label', kde=True, palette='coolwarm')
plt.title('Distribution of Risk Scores by Group')
plt.xlabel('Risk Score')
plt.grid(True)
plt.show()
```

```python
!pip install ipywidgets --quiet

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets, VBox, HBox

# For better plots in Colab
%matplotlib inline

# Load your cleaned dataset (already processed)
df = pd.read_csv('/content/healthcare_dataset.csv')  # Adjust if needed

# Display some basic info
df.head()
```

```python
# Set ranges for Age and Billing
age_slider = widgets.IntRangeSlider(
    value=[20, 60],
    min=int(df['Age'].min()),
    max=int(df['Age'].max()),
    step=1,
    description='Age Range:',
    continuous_update=False
)

billing_slider = widgets.FloatRangeSlider(
    value=[df['Billing Amount'].min(), df['Billing Amount'].max()],
    min=float(df['Billing Amount'].min()),
    max=float(df['Billing Amount'].max()),
    step=10.0,
    description='Billing Amt:',
    continuous_update=False
)

risk_dropdown = widgets.Dropdown(
    options=['All', 'Low Risk', 'Medium Risk', 'High Risk'],
    value='All',
    description='Risk Level:'
)
```

```python
def show_filtered_data(age_range, billing_range, risk_level):
    # Filter dataset
    filtered_df = df[
        (df['Age'] >= age_range[0]) & (df['Age'] <= age_range[1]) &
        (df['Billing Amount'] >= billing_range[0]) & (df['Billing Amount'] <= billing_range[1])
    ]

    if risk_level != 'All':
        filtered_df = filtered_df[filtered_df['Risk Label'] == risk_level]

    # Display summary
    print(f"ðŸ‘¥ Employees in selection: {len(filtered_df)}")
    display(filtered_df[['Age', 'Billing Amount', 'Medical Condition', 'Risk Score', 'Risk Label', 'Intervention']].head())

    # Plot Risk Score distribution
    plt.figure(figsize=(10, 5))
    sns.histplot(filtered_df['Risk Score'], kde=True, color='steelblue')
    plt.title('Filtered Risk Score Distribution')
    plt.xlabel('Risk Score')
    plt.grid(True)
    plt.show()
```

```python
# Encode text fields into numeric categories
df['Medical Condition'] = df['Medical Condition'].astype('category').cat.codes
df['Admission Type'] = df['Admission Type'].astype('category').cat.codes
df['Medication'] = df['Medication'].astype('category').cat.codes

# Map test results (if not numeric yet)
if df['Test Results'].dtype == 'object':
    df['Test Results'] = df['Test Results'].map({
        'Normal': 0, 'Abnormal': 1, 'Inconclusive': 0.5
    })
```

```python
from sklearn.preprocessing import StandardScaler

features = ['Age', 'Medical Condition', 'Billing Amount', 'Admission Type', 'Medication', 'Test Results']
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[features])
```

```python
# Re-import if needed
from sklearn.ensemble import IsolationForest
from sklearn.cluster import KMeans

# Recalculate if not done
features = ['Age', 'Medical Condition', 'Billing Amount', 'Admission Type', 'Medication', 'Test Results']
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[features])

# Anomaly Detection
iso = IsolationForest(contamination=0.05, random_state=42)
df['Risk Score'] = -iso.fit(df_scaled).decision_function(df_scaled)

# Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df['Risk Group'] = kmeans.fit_predict(df[['Risk Score']])
```

```python
# Map cluster numbers to readable labels (adjust if needed)
cluster_map = {
    df.groupby('Risk Group')['Risk Score'].mean().idxmin(): 'Low Risk',
    df.groupby('Risk Group')['Risk Score'].mean().median(): 'Medium Risk',
    df.groupby('Risk Group')['Risk Score'].mean().idxmax(): 'High Risk'
}

df['Risk Label'] = df['Risk Group'].map(cluster_map)
```

```python
print(df[['Risk Score', 'Risk Group', 'Risk Label']].head())
```

```python
import numpy as np

# Precompute threshold to avoid recalculating
high_risk_score_threshold = df['Risk Score'].quantile(0.95)

# Define conditions
conditions = [
    df['Risk Score'] > high_risk_score_threshold,
    df['Risk Label'] == 'High Risk',
    df['Risk Label'] == 'Medium Risk'
]

# Define corresponding outputs
choices = [
    'Immediate Diagnostic Follow-up',
    'Weekly Health Counseling',
    'Monthly Monitoring'
]

# Default intervention if none of the above matched
default = 'Quarterly Wellness Tips'

# Apply the conditions
df['Intervention'] = np.select(conditions, choices, default=default)
```
